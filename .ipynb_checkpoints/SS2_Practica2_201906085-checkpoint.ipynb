# Practica 2 - Análisis de datos con Python

```python
import pandas as pd

# Leer el archivo CSV
datos = pd.read_csv("datos.csv", names=[
    "Course Title",
    "Rating",
    "Level",
    "Duration",
    "Schedule",
    "Review",
    "What you will learn",
    "Skill gain",
    "Modules",
    "Instructor",
    "Offered By",
    "Keyword",
    "Course Url"
])

# Función para eliminar corchetes y comillas simples
def limpiar_datos(valor):
    if isinstance(valor, str):
        valor = valor.replace("[", "").replace("]", "").replace("'", "").replace("\"","")
    return valor

# Aplicar la función a las columnas especificadas
columnas_a_limpiar = ["What you will learn", "Skill gain", "Modules", "Instructor", "Offered By", "Keyword"]
datos[columnas_a_limpiar] = datos[columnas_a_limpiar].map(limpiar_datos)

# Convertir la columna 'Rating' a tipo float, manejar errores
def convertir_a_float(valor):
    try:
        return float(valor)
    except ValueError:
        return None

datos['Rating'] = datos['Rating'].map(convertir_a_float)

# Eliminar filas con valor NaN en la columna 'Rating'
datos = datos.dropna(subset=['Rating'])

# Calcular el promedio de calificaciones para cada curso
promedio_calificaciones = datos.groupby('Course Title')['Rating'].mean()

# Mostrar todos los promedios de calificaciones
#pd.set_option('display.max_rows', None)
print("Promedio de calificaciones por curso:", round(promedio_calificaciones,2))



# Restaurar la configuración de visualización
pd.reset_option('display.max_rows')

# Cursos con la calificación más alta
max_rating = promedio_calificaciones.max()
cursos_max_rating = promedio_calificaciones[promedio_calificaciones == max_rating].index.tolist()
print("-------------Cursos con mayor rating:-------------\n\n", cursos_max_rating, "- Rating:", max_rating)
print()

# Cursos con la calificación más baja
min_rating = promedio_calificaciones.min()
cursos_min_rating = promedio_calificaciones[promedio_calificaciones == min_rating].index.tolist()
count=0
cadena_curso_min = ""
for cursos_min in cursos_min_rating:
    count= count +1 
    cadena_curso_min += str(count) + ". " + cursos_min +"\n"
    
print("-------------Cursos con menor rating:-------------\n\n", cadena_curso_min, "- Rating:", min_rating)
print()

# Calcular el porcentaje de cursos con horario flexible
total_cursos = datos['Course Title'].nunique()
cursos_horario_flexible = datos[datos['Schedule'] == 'Flexible schedule']['Course Title'].nunique()
porcentaje_horario_flexible = (cursos_horario_flexible / total_cursos) * 100
porcentaje_horario_flexible = round(porcentaje_horario_flexible, 2)

print("Porcentaje de cursos con horario flexible:", porcentaje_horario_flexible, "%")
```
